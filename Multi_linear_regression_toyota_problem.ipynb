## 4) Model Building :

```python
X = toyota_dataset_1.drop(labels ='price', axis = 1)
y = toyota_dataset_1[['price']]
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
scaled_X= std_scaler.fit_transform(X)
scaled_X
```

```python
scaled_X = pd.DataFrame(data=scaled_X,columns=['age','km','hp','cc','doors','gears','quarterly_tax','weight'])
scaled_X
```

## 5) Model Training :

```python
from sklearn.linear_model import LinearRegression
linear_model_1 = LinearRegression()
linear_model_1.fit(scaled_X,y)
```

## 6) Model  Testing :

```python
y_predict = linear_model_1.predict(scaled_X)
y_predict
```

## 7) Model Evaluation :

```python
error = y - y_predict
error
```

### Model building :

```python
import statsmodels.formula.api as smf
```

```python
model_1 = smf.ols(formula = 'price ~ age + km + hp + cc + doors + gears + quarterly_tax + weight',data=toyota_dataset_1).fit()
model_1
```

```python
y_predict_1 = model_1.predict(X)
```

#### Model Testing :

```python
model_1.params
```

```python
model_1.pvalues
```

```python
model_1.tvalues
```

```python
print('R2 Score          :',round(model_1.rsquared,5)) 
print('Adjusted R2 Score :',round(model_1.rsquared_adj,5))
print('AIC               :',round(model_1.aic,4))
print('BIC               :',round(model_1.bic,4))
```

```python
model_1.summary()
```

### Model Deletion Diagnostic

#### Detecting Outliners and influencers

#### Cook's Distance

```python
from statsmodels.graphics.regressionplots import influence_plot
```

```python
model_1_influence = model_1.get_influence()
(c,_) = model_1_influence.cooks_distance
```

```python
#Plotting influencers

fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(toyota_dataset_1)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#Checking the index and c value of influencer

(np.argmax(c),np.max(c))
```

```python
#High influence plot

influence_plot(model_1)
plt.show()
```
